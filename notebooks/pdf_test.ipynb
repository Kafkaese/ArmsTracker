```python
from tabula.io import read_pdf
def german_report(pdf_path):
    '''
    Extract table of exports from bi-annual report of BMWK
    '''
    
    report = read_pdf(pdf_path, pages='all')
    
    return report
```

```python
tables = german_report('../taro/data/Ruestungsexportbericht-2018.pdf')
```

```python
total_per_country = tables[90]
```
